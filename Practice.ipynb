```python
import statistics as stats
#1. Open the file with the open()
control_data = open("data/control.txt")
#2. Read the file's lines with .readlines(). 
    #this returns a list of strings, one string
    #per line, each ending with'\n'
read_control_data = control_data.readlines()
#print(read_control_data)
#3. Use the .strip() to remove trailing spaces
    # and the '\n' character
clean_control_data = []
for line in read_control_data:
    line.strip()
    if not line.strip() == 'minutes':
        clean_control_data.append(float(line))

print(clean_control_data)

stats(clean_control_data) 
  
```

```python
file_name = "data/phase0.txt"
file = open(file_name)
lines = file.readlines()

clean_data = []
for line in lines:
    if line.strip().isdigit():
        clean_data.append(int(line))

    
print(stats.std(clean_data))
```
